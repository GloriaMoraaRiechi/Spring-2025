```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, regexp_replace, split, explode
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF
from pyspark.ml.classification import DecisionTreeClassifier, RandomForestClassifier, NaiveBayes
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
import matplotlib.pyplot as plt
```

Initialize spark session

```python
spark = SparkSession.builder.appName("Covid19_Sentiment_Analysis").getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive')

# Provide the actual file path
file_path = "/content/drive/My Drive/Dataset/training.csv"

train_df = spark.read.csv(file_path, inferSchema=True, header=True)
train_df.show()
```

# Data Preprocessing

Rename columns

```python
train_df = train_df.withColumnRenamed("Tweet", "text").withColumnRenamed("Labels", "labels")
```

```python
def preprocess_text(df):
    df = df.withColumn("text", regexp_replace(col("text"), r"http\S+", ""))  # Remove URLs
    df = df.withColumn("text", regexp_replace(col("text"), r"@\W+", ""))  # Remove mentions
    df = df.withColumn("text", regexp_replace(col("text"), r"[^a-zA-Z ]", ""))  # Remove special characters
    return df

train_df = preprocess_text(train_df)
```

```python
train_df.show()
```

Tokenization and Stopword Removal

```python
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF
```

```python
# Step 1: Tokenize the text
tokenizer = Tokenizer(inputCol="text", outputCol="tokenized_words")
```

```python
# Step 2: Remove stopwords
remover = StopWordsRemover(inputCol="tokenized_words", outputCol="filtered")
```

```python
# Step 3: Apply HashingTF
hashingTF = HashingTF(inputCol="filtered", outputCol="rawFeatures", numFeatures=10000)
```

```python
idf_model = IDF(inputCol="rawFeatures", outputCol="features")
```

```python
train_df.printSchema()
```

# Model Training

Explode labels

explode() will transform the array columns into multiplw rows where each element in the array becomes its own row

```python
# Step 1: Split the 'labels' string into an array
train_df = train_df.withColumn("labels", split(col("labels"), r"\s+"))

# Step 2: Explode the 'labels' array into separate rows
train_df = train_df.withColumn("label", explode(col("labels")))

# Step 3: Cast the 'label' column to integer
train_df = train_df.withColumn("label", col("label").cast("int"))

train_df.show()
```

```python
train_data, test_data = train_df.randomSplit([0.8, 0.2], seed=42)
```

Classifiers

```python
classifiers = {
    "DecisionTree": DecisionTreeClassifier(featuresCol="features", labelCol="label"),
    "RandomForest": RandomForestClassifier(featuresCol="features", labelCol="label", numTrees=50, maxDepth=10),
    "NaiveBayes": NaiveBayes(featuresCol="features", labelCol="label")
}
```

Evaluators

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix, precision_score, recall_score, f1_score, accuracy_score

accuracy_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
f1_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="f1")
precision_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="precision")
recall_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="recall")
```

Model training and evaluation

```python
from sklearn.metrics import roc_curve, auc, precision_score, recall_score, confusion_matrix
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np

results = {}

for name, classifier in classifiers.items():
    print(f"Training {name}...")

    # Define the pipeline stages
    pipeline = Pipeline(stages=[tokenizer, remover, hashingTF, idf_model, classifier])

    # Fit the pipeline to the training data
    model = pipeline.fit(train_data)

    predictions = model.transform(test_data)

    accuracy = accuracy_evaluator.evaluate(predictions)
    f1 = f1_evaluator.evaluate(predictions)

    # Convert predictions to pandas for sklearn metrics
    y_true = predictions.select("label").toPandas()
    y_pred = predictions.select("prediction").toPandas()

    precision = precision_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')

    results[name] = {
        "accuracy": accuracy,
        "f1_score": f1,
        "precision": precision,
        "recall": recall
    }

    print(f"{name} - Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}")

    num_classes = len(np.unique(y_true))

    # ROC and AUC for multiclass
    y_true_binarized = label_binarize(y_true, classes=[i for i in range(num_classes)])  # Binarize labels
    y_pred_proba = predictions.select("probability").toPandas()['probability'].apply(lambda x: np.array(x.toArray()))

    plt.figure(figsize=(10, 8))
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_binarized[:, i], [prob[i] for prob in y_pred_proba])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'Class {i} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for {name}')
    plt.legend(loc="lower right")
    plt.show()

    cm = confusion_matrix(y_true, y_pred)
    print(f"Confusion Matrix for {name}:")
    print(cm)

for model, scores in results.items():
    print(f"\n🔹 {model} Performance:")
    for metric, value in scores.items():
        print(f"  {metric}: {value:.4f}")
```
